# Implemented Kmeans

```python
import numpy as np
import matplotlib.pyplot as plt
import math
import random
from random import randint
from sklearn.metrics.pairwise import rbf_kernel
from numpy.linalg import inv 
from numpy import linalg as LA
from sklearn.cluster import KMeans

def cv2_imshow(a):
  """A replacement for cv2.imshow() for use in Jupyter notebooks."""
  a = a.clip(0, 255).astype('uint8')
  # cv2 stores colors as BGR; convert to RGB
  if a.ndim == 3:
    if a.shape[2] == 4:
      a = cv2.cvtColor(a, cv2.COLOR_BGRA2RGBA)
    else:
      a = cv2.cvtColor(a, cv2.COLOR_BGR2RGB)
  display(PIL.Image.fromarray(a))
  
def Euclidean(a,b):
    return math.sqrt((a[0]-b[0])**2 +(a[1]-b[1])**2+(a[2]-b[2])**2)
    
def Kmeans(data,k):
    
    n = len(data)
    rand = [i for i in range(0,n)]
    random.shuffle(rand)
    means = np.array([data[rand[i]] for i in range(0,k)] , dtype=np.float)
    old_means = np.array([data[rand[i]] for i in range(0,k)] , dtype=np.float)
    iterations = 0
    
    while iterations < 20:
        iterations += 1

        labels = [0]*n

        for i in range(0,n):
            # for this point find its label
            cur = -1
            idx = 0

            for j in range(0,k):
                tmp = Euclidean(data[i],means[j])
                if cur == -1 or tmp < cur :
                    idx = j
                    cur = tmp

            labels[i] = idx

        for i in range(0,k):
            sum = 0
            point = [0.0,0.0,0.0]
            for j in range(0,n):
                if labels[j] == i:
                    sum += 1
                    point[0] += data[j][0]
                    point[1] += data[j][1]
                    point[2] += data[j][2]
            
            if sum != 0:
                means[i] = [point[0]/sum , point[1]/sum, point[2]/sum]
             
#         if np.array_equal(means,old_means) :
#             break

        old_means = np.array([means[i] for i in range(0,k)] , dtype=np.float)
    
    newImage = []
    for i in range(0,len(data)):
        newImage.append(old_means[labels[i]])
   
    newImage = np.array(newImage).reshape(481,321,3)
    cv2_imshow(newImage)      
    colors = [[192,192,192],[0,0,0],[255,255,255],[255,51,51],[102,204,0],[0,255,255],[96,96,96],[51,51,255],[204,255,255],[200,200,200],[100,100,100]]
    c = []
    for i in range(0,len(data)):
        c.append(colors[labels[i]])
    newImage = np.array(c).reshape(481,321,3)
    cv2_imshow(newImage)     
        
       
        
for i in range(1,6):
  img = cv2.imread('real'+str(i)+'.jpg')
  data = np.array(img).reshape(481*321,3)
  Kmeans(data,5)
  
```
